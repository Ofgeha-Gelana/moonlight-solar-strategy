```python
# Visualize outliers in wind speed data

# Drop non-numeric columns (like 'Comments') for the quality check
numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(14, 6))
sns.boxplot(data=numeric_df[['WS', 'WSgust']], orient='v')
plt.title('Outlier Detection in Wind Speed Data')
plt.show()
```
