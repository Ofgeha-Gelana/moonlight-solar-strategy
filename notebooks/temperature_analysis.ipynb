```python
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
import pandas as pd
```

```python
rpath = os.path.abspath('..')
if rpath not in sys.path:
    sys.path.insert(0, rpath)
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv", parse_dates=['Timestamp'])
    # Set the Timestamp as index for time series plotting
df.set_index('Timestamp', inplace=True)
```

```python
df.columns
```

```python
# Scatter plot for temperature vs. RH
plt.figure(figsize=(8, 6))
sns.scatterplot(x='RH', y='Tamb', data=df)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# Scatter plot for solar radiation vs. RH
plt.figure(figsize=(8, 6))
sns.scatterplot(x='RH', y='WSstdev', data=df)
plt.title('Solar Radiation vs. Relative Humidity')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Solar Radiation (W/m²)')
plt.show()
```
