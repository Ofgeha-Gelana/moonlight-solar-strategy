```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys
```

```python
rpath = os.path.abspath('..')
if rpath not in sys.path:
    sys.path.insert(0, rpath)
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv", parse_dates=['Timestamp'])
    # Set the Timestamp as index for time series plotting
df.set_index('Timestamp', inplace=True)
```

```python
benin_file_path = "../data/raw/benin-malanville.csv"
sierraleone_file_path = "../data/raw/sierraleone-bumbuna.csv"
togo_file_path = "../data/raw/togo-dapaong_qc.csv"
```

```python
# importing load_file function from scripts
from scripts.loading_data import load_file

benin_df = load_file(benin_file_path)
sierraleone_df = load_file(sierraleone_file_path)
togo_df = load_file(togo_file_path)
```

```python
# Creating histograms for a list of specified variables in the given DataFrame.
from scripts.plot_histograms import plot_histograms
variables = ['GHI', 'DNI', 'DHI', 'WS', 'Tamb', 'TModA', 'TModB']
print("Histograms for Benin dataset")
plot_histograms(benin_df, variables)

print("Histograms for sierralone dataset")
plot_histograms(sierraleone_df, variables)

print("Histograms for togo dataset")
plot_histograms(togo_df, variables)
```

#### Histogram for GHI (Global Horizontal Irradiance)



```python
# Histogram for GHI
plt.figure(figsize=(8, 6))
sns.histplot(df['GHI'], bins=30, kde=True)
plt.title('Histogram of GHI (Global Horizontal Irradiance)')
plt.xlabel('GHI (W/m²)')
plt.ylabel('Frequency')
plt.show()
```

#### Histogram for DNI (Direct Normal Irradiance)



```python
# Histogram for DNI
plt.figure(figsize=(8, 6))
sns.histplot(df['DNI'], bins=30, kde=True)
plt.title('Histogram of DNI (Direct Normal Irradiance)')
plt.xlabel('DNI (W/m²)')
plt.ylabel('Frequency')
plt.show()

```

#### Histogram for DHI (Diffuse Horizontal Irradiance):



```python
# Histogram for DHI
plt.figure(figsize=(8, 6))
sns.histplot(df['DHI'], bins=30, kde=True)
plt.title('Histogram of DHI (Diffuse Horizontal Irradiance)')
plt.xlabel('DHI (W/m²)')
plt.ylabel('Frequency')
plt.show()
```

#### Histogram for Wind Speed (WS):



```python
# Histogram for Wind Speed (WS)
plt.figure(figsize=(8, 6))
sns.histplot(df['WS'], bins=30, kde=True)
plt.title('Histogram of Wind Speed (WS)')
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('Frequency')
plt.show()
```
